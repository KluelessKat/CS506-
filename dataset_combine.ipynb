```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#concatenate 2020-2022 datasets

data1 = pd.read_stata("2022.dta")
data2 = pd.read_stata("2021.dta")
data3 = pd.read_stata("2020.dta")

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

# Step 2: Define the list of required columns to keep
required_columns = [
    'REGION22',   # Region of residence
    'FAMS1231',   # Family size
    'RXEXP22',    # Prescription expenses
    'CANCERDX',   # Cancer diagnosis
    'UNINS22',    # Uninsured status
    'PRVEV22',    # Private insurance coverage
    'MCDEV22',    # Medicaid eligibility
    'MCREV22',    # Medicare eligibility
    'HOUR53',     # Hours worked
    'EMPST53',    # Employment status
    'FAMINC22',   # Family income
    'HIDEG',      # Highest degree obtained
    'MARRY22X',   # Marital status
    'RACEV1X',    # Race/ethnicity
    'SEX',        # Gender
    'AGELAST',    # Age of the individual
    'TOTSLF22',   # Total self-pay medical expenses
    # Cancer-specific diagnosis columns
    'CABLADDR',   # Bladder cancer
    'CABREAST',   # Breast cancer
    'CACERVIX',   # Cervical cancer
    'CACOLON',    # Colon cancer
    'CALUNG',     # Lung cancer
    'CALYMPH',    # Lymphoma
    'CAMELANO',   # Melanoma
    'CAOTHER',    # Other cancer
    'CAPROSTA',   # Prostate cancer
    'CASKINNM',   # Non-melanoma skin cancer
    'CASKINDK',   # Unknown skin cancer type
    'CAUTERUS'    # Uterine cancer
]

# Step 2: Filter each dataset to retain only columns in the required list
def filter_to_required_columns(df, required_columns):
    # Identify columns that exist in both the dataset and the required columns
    valid_columns = [col for col in df.columns if col in required_columns]
    # Filter the dataset to these columns
    return df[valid_columns]

df1_filtered = filter_to_required_columns(df1, required_columns)
df2_filtered = filter_to_required_columns(df2, required_columns)
df3_filtered = filter_to_required_columns(df3, required_columns)

# Step 3: Concatenate the filtered datasets
combined_df = pd.concat([df1_filtered, df2_filtered, df3_filtered], ignore_index=True)

# Step 4: Print columns in the combined dataset to verify
print(f"Columns in the combined dataset: {list(combined_df.columns)}")

# Print num rows and columns
print(f'Number of rows: {combined_df.shape[0]}')
print(f'Number of columns: {combined_df.shape[1]}')

print(combined_df.head())
```

```python
# Preprocess row values

# Transformation mappings for specific columns
transformations = {
    'SEX': lambda x: x.astype(str).str.split().str[1] if x.dtype != object else x.str.split().str[1],  # Extract the second word (e.g., "1 MALE" -> "MALE") # Convert to string before splitting if not already string
    'RACEV1X': lambda x: x.apply(
        lambda race: 'WHITE' if 'WHITE' in race else
                     'BLACK' if 'BLACK' in race else
                     'AMER INDIAN/ALASKA NATIVE' if 'AMER INDIAN' in race or 'ALASKA NATIVE' in race else
                     'ASIAN' if 'ASIAN' in race or 'NATV HAWAIIAN' in race or 'PACFC ISL' in race else
                     'MULTIPLE' if 'MULTIPLE' in race else
                     'UNKNOWN'),
    'MARRY22X': {
        '2 WIDOWED': 'WIDOWED',
        '3 DIVORCED': 'DIVORCED',
        '5 NEVER MARRIED': 'NEVER MARRIED',
        '6 UNDER AGE 16 - INAPPLICABLE': 'UNDER AGE 16',
        '1 MARRIED': 'MARRIED',
        '4 SEPARATED': 'SEPARATED',
        '-7 REFUSED': 'REFUSED',
        '-8 DON\'T KNOW': 'DON\'T KNOW'
    },
    'HIDEG': {
        '1 NO DEGREE': 'NO DEGREE',
        '7 OTHER DEGREE': 'OTHER DEGREE',
        '5 MASTER\'S DEGREE': 'MASTER\'S DEGREE',
        '4 BACHELOR\'S DEGREE': 'BACHELOR\'S DEGREE',
        '8 UNDER AGE 16 - INAPPLICABLE': 'UNDER AGE 16',
        '3 HIGH SCHOOL DIPLOMA': 'HIGH SCHOOL DIPLOMA',
        '2 GED': 'GED',
        '6 DOCTORATE DEGREE': 'DOCTORATE DEGREE',
        '-8 DON\'T KNOW': 'DON\'T KNOW',
        '-7 REFUSED': 'REFUSED'
    },
    'EMPST53': {
        '4 NOT EMPLOYED DURING RD 5/3': 'NOT EMPLOYED',
        '3 JOB DURING RD 5/3 REF PERIOD': 'JOB DURING REF',
        '1 EMPLOYED AT RD 5/3 INT DATE': 'EMPLOYED',
        '-1 INAPPLICABLE': 'INAPPLICABLE',
        '2 JOB TO RETURN TO AT RD 5/3 INT DATE': 'HAS JOB TO RETURN',
        '-7 REFUSED': 'REFUSED',
        '-15 CANNOT BE COMPUTED': 'CANNOT COMPUTE',
        '-8 DON\'T KNOW': 'DON\'T KNOW'
    },
    'REGION22': {
        '-1 INAPPLICABLE': 'INAPPLICABLE',
        '1 NORTHEAST': 'NORTHEAST',
        '2 MIDWEST': 'MIDWEST',
        '3 SOUTH': 'SOUTH',
        '4 WEST': 'WEST'
    }
}

# Columns with Yes/No values
yes_no_columns = ['PRVEV22', 'MCREV22', 'MCDEV22', 'UNINS22', 'CANCERDX',
                  'CABLADDR', 'CABREAST', 'CACERVIX', 'CACOLON',
                  'CALUNG', 'CALYMPH', 'CAMELANO', 'CAOTHER',
                  'CAPROSTA', 'CASKINNM', 'CASKINDK', 'CAUTERUS']

print(combined_df[yes_no_columns].head())

# Invalid numeric values for conversion
invalid_values = ['-15 CANNOT BE COMPUTED', '-8 DON\'T KNOW', '-7 REFUSED', '-2 DETERMINED IN PREVIOUS ROUND']

def preprocess_columns(df):
    # Apply transformations for specific columns
    for column, rule in transformations.items():
        if column in df.columns:
            if callable(rule):
                df[column] = rule(df[column])  # Apply function-based transformation
            elif isinstance(rule, dict):
                df[column] = df[column].replace(rule)  # Apply mapping-based transformation

    # Process Yes/No columns
    for column in yes_no_columns:
        if column in df.columns:
            df[column] = df[column].str.split().str[0]

    # Convert numeric columns with invalid values to NaN
    numeric_columns = ['HOUR53', 'FAMS1231', 'AGELAST', 'TOTSLF22', 'RXEXP22']
    for column in numeric_columns:
        if column in df.columns:
            df[column] = df[column].replace(invalid_values, np.nan)
            df[column] = pd.to_numeric(df[column], errors='coerce')

    # Filter rows where CANCERDX is not 'YES'
    if 'CANCERDX' in df.columns:
        df = df[df['CANCERDX'] == '1']

    return df

# Preprocess the combined dataset
combined_df = preprocess_columns(combined_df)

# Save the combined dataset
combined_df.to_csv("combined_2020-2022.csv", index=False)

# Print the first few rows
print(combined_df.head())

# Print num rows and columns
print(f'Number of rows: {combined_df.shape[0]}')
print(f'Number of columns: {combined_df.shape[1]}')
```

```python
# Categorical Variables Distribution
df = combined_df

def plot_categorical_distributions(df):
    categorical_cols = ['SEX', 'RACEV1X', 'MARRY22X', 'HIDEG', 'EMPST53', 'REGION22', 'MCREV22', 'MCDEV22', 'PRVEV22', 'UNINS22', 'CANCERDX']

    fig, axes = plt.subplots(len(categorical_cols), 1, figsize=(10, 4*len(categorical_cols)))
    fig.suptitle('Distribution of Categorical Variables', fontsize=16)

    for i, col in enumerate(categorical_cols):
        df[col].value_counts().plot(kind='bar', ax=axes[i], title=col)
        axes[i].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

plot_categorical_distributions(df)



# Numerical Variables Correlation Heatmap
def plot_correlation_heatmap(df):

    plt.figure(figsize=(12, 10))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Heatmap of Numerical Variables')
    plt.tight_layout()
    plt.show()

plot_correlation_heatmap(df)


def plot_barplots(df):
    categorical_cols = ['SEX', 'RACEV1X', 'MARRY22X', 'HIDEG', 'EMPST53', 'REGION22']
    categorical_col_titles = ['Sex', 'Race', 'Marital Status', 'Highest Education', 'Employment Status', 'Region']
    fig, axes = plt.subplots(len(categorical_cols), 1, figsize=(10, 4*len(categorical_cols)))
    fig.suptitle('TOTSLF22 Mean by Categorical Variables', fontsize=16)

    for i, col in enumerate(categorical_cols):
        sns.barplot(x=col, y='TOTSLF22', data=df, ax=axes[i], ci=None)
        axes[i].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

plot_barplots(df)

```
